```python
import pandas as pd 
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
df1 = pd.read_csv("movies.csv")
```

```python
selected_features = ['genres','tagline','keywords','cast','director']
```

```python
for feature in selected_features:
    df1[feature] = df1[feature].fillna('')
```

```python
df1.isnull().sum()
```

```python
# combined feauture for vectorization
```

```python
combinded_features = df1['genres']+' '+df1['keywords']+' '+df1['tagline']+' '+df1['cast']+' '+df1['director']
```

```python
combinded_features
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combinded_features)
```

```python
print(feature_vectors)
```

```python
# now use cosine similarity for looking for similar looking value.
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```
